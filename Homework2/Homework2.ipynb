## 5.1 线性回归建模

```python
quali_vars=['nature','size','sector','education','experience']
dummy1 = pd.get_dummies(data[quali_vars])#对多属性取值的自变量做哑变量处理
dummy1
```

```python
dummy1.drop(labels=['nature_国企','size_20人以下','sector_银行','education_本科','experience_1_3年'],axis=1)
#选取“国企”，“20人以下”，“银行”，“本科”，“1-3年”做基准组，故删除这些列
```

```python
newdata = pd.concat([data,dummy1],axis=1)#将处理后的哑变量和原数据拼接
newdata.drop(quali_vars,axis=1,inplace=True)#删除原始自变量数据
order=[#因变量
    'lnsalary',
    #定性变量
    #1.公司属性，基准组：国企
    'nature_民营','nature_股份制企业','nature_上市公司','nature_合资','nature_外商独资','nature_其他',
    #2.公司规模，基准组：20人以下
    'size_20-99人','size_100-499人','size_500-999人','size_1000-9999人','size_10000人以上',
    #3.公司行业大类，基准组：银行
    'sector_基金,证券,期货,投资','sector_信托,担保,拍卖,典当','sector_保险',
    #4.学历，基准组：本科
    'education_不限','education_大专','education_硕士及以上',
    #5.经验，基准组：1-3年
    'experience_不限','experience_无经验','experience_1年以下','experience_3_5年','experience_5_10年','experience_10年以上'
]
    
newdata=newdata[order]#调整列顺序
```

```python
#y:lnsalary,x1:民营,x2:股份制企业,x3:上市公司,x4:合资,x5:外商独资,x6:属性为其他,x7:20~99人,x8:100~499人,x9:500~999人,x10:1000~9999人,x11:10000人以上,x12:基金证券期货投资,x13:信托担保拍卖典当,x14:保险,x15:教育不限,x16:大专,x17:硕士及以上,x18:经验不限,x19:无经验,x20:1年以下,x21:3`5年,x22:5`10年,x23:10年以上
newdata.columns=["lnsalary","x1","x2","x3","x4","x5","x6","x7","x8","x9","x10","x11",
                "x12","x13","x14","x15","x16","x17","x18","x19","x20","x21","x22","x23"]
newdata
```

```python
X = newdata[newdata.columns[1:]]
Y = newdata['lnsalary']
X = sm.add_constant(X)
est = sm.OLS(Y,X).fit()
est.summary()#线性回归建模
```

## 6. 利用BIC准则进行变量选择及使用五折交叉验证

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
def backward(df):#定义逐步回归函数，使用bic准则
    all_bianliang = [i for i in range(0,23)]#备退因子
    ceshi = [i for i in range(0,23)]#存放加入单个因子后的模型
    zhengshi = [i for i in range(0,23)]#收集确定因子
    data1 = pd.concat([df.iloc[:,i+1] for i in ceshi ],axis = 1)#结合所需因子
    name = 'lnsalary~'+'+'.join(data1.columns)
    result = smf.ols(name,data=df).fit()#建模
    c0 = result.bic #最小bic
    delete = []#已删元素
    while(all_bianliang):
        bic = []#存放bic
 
        for i in all_bianliang:
            ceshi = [i for i in zhengshi]
            ceshi.remove(i)
            data1 = pd.concat([df.iloc[:,i+1] for i in ceshi ],axis = 1)#结合所需因子
            name = "lnsalary~"+("+".join(data1.columns))#组成公式
            data = pd.concat([df['lnsalary'],data1],axis=1)#结合自变量和因变量
            result = smf.ols(name,data=data).fit()#建模
            bic.append(result.bic)#将所有bic存入
 
        if min(bic)>c0:#bic已经达到最小
            data1 = pd.concat([df.iloc[:,i+1] for i in zhengshi ],axis = 1)#结合所需因子
            name = "lnsalary~"+("+".join(data1.columns))#组成公式
            break
            
        else:
            zhengshi.remove(all_bianliang[bic.index(min(bic))])#查找最小的bic并将最小的因子存入正式的模型列表当中
 
        c0 = min(bic)
        delete.append(bic.index(min(bic)))
        all_bianliang.remove(all_bianliang[delete[-1]])#删除已删因子
 
    name = "lnsalary~"+("+".join(data1.columns))#组成公式
    print("最优模型为：{}，其bic为：{}".format(name,c0))
    result = smf.ols(name,data=df).fit()#建模
    print()
    print(result.summary())
    print(result.params)


```

```python
backward(newdata)
```

回归模型经BIC准则进行变量选择后的结果为：  
1. 截距项：其回归系数值为9.06
2. 在公司属性变量中，选择民营,股份制企业,上市公司,合资,外商独资,属性为其他
3. 在公司规模变量中，选择了10000人以上的公司
4. 在公司行业大类变量中，选择了信托担保拍卖典当,保险
5. 在学历变量中，选择了教育不限,大专,硕士及以上
6. 在经验变量中，选择了无经验，3-5年，5-10年，10年以上

其可能的解释为，公司属性对于薪资的影响比较大，所以属性这一变量的所有取值均被选择；对于公司规模，10000人以上的公司的薪资平均水平最高，影响薪资的变化最大，其余取值对薪资的影响较小，所以仅仅选择了10000人以上的公司；对于公司行业大类，信托担保拍卖典当和保险这两个大类对于薪资的影响水平更高；对于学历，教育不限，大专，硕士及以上均是对薪资影响水平很高的取值，其薪资水平均比基准组“本科”差距较大；对于经验，无经验，3-5年，5-10年，10年以上对于薪资的影响更大，所以入选。综上，经BIC准则选择的变量均是对因变量影响较大的变量取值。

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
```

```python
linreg = LinearRegression()
linreg.fit(newdata[newdata.columns[1:]], newdata['lnsalary'])
crossvalidation = KFold(n_splits=5, random_state=1, shuffle=True)
scores = cross_val_score(linreg, X, Y, scoring="neg_mean_squared_error", cv=crossvalidation, n_jobs=1)#进行五折交叉验证，评分标准为均方根误差
mae = np.abs(scores.mean())
rmse = np.sqrt(-scores)
print("\nLinear Regression Performance on the test set: RMSE (Root Mean Squared Error) : {:.4f} \n".format(rmse.mean()))
print("Linear Regression Performance on the test set: MAE (Mean Absolute Error) : %.3f" %(mae))
```

模型经变量选择，五折交叉验证后，RMSE和MAE均较小，说明模型拟合得比较好。综上，线性回归模型在对各属性做哑变量处理，经逐步回归，以BIC准则筛选变量后得到的自变量均是对因变量水平影响较大的变量，其能很好的解释因变量的变化，得到一个很好的拟合结果。
