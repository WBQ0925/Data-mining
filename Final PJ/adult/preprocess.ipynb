```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from category_encoders import TargetEncoder

# 读取数据
df = pd.read_csv('adult.csv')

# 处理缺失值
df[df == '?'] = np.nan
for col in ['workclass', 'occupation', 'native.country']:
    df[col].fillna(df[col].mode()[0], inplace=True)
 

X = df.drop(['income'], axis=1)
y = df['income']

# 0-1标签
y = (y == '>50K').astype(int)
X['sex'] = (X['sex'] == 'Male').astype(int)


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

 # label编码
# categorical = ['workclass', 'education', 'marital.status', 'occupation', 'relationship', 'race', 'native.country']
# for feature in categorical:
#         le = preprocessing.LabelEncoder()
#         X_train[feature] = le.fit_transform(X_train[feature])
#         X_test[feature] = le.transform(X_test[feature])


# target 编码 # 注意，目标编码会用到y，因此我们只能用训练集fit
categorical = ['workclass', 'education', 'marital.status', 'occupation', 'relationship', 'race', 'native.country']

enc = TargetEncoder(cols=categorical, min_samples_leaf=20, smoothing=10).fit(X_train, y_train)
X_train = enc.transform(X_train)
X_test = enc.transform(X_test)



# 标准化
scaler = StandardScaler()

X_train = pd.DataFrame(scaler.fit_transform(X_train), columns = X.columns, index = X_train.index)

X_test = pd.DataFrame(scaler.transform(X_test), columns = X.columns, index = X_test.index)

# 统一后保存
X_train['label'] = y_train

X_test['label'] = y_test

# X_train.to_csv('train1.csv')
# X_test.to_csv('test1.csv')
```
